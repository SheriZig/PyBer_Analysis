```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt() object-oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis,  marker="d", color="green", linewidth=2,label = 'Boston')

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

#Add the legend
plt.legend()
```

```python
#This is a typed in dataset
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Adding 'o' inside parentheses to change the plot from a line to a scatter plot 
plt.plot(x_axis, y_axis, 'o')
```

```python
#Create scatter plot using the plt.scatter() function
plt.scatter(x_axis,y_axis)
```

```python
#formatting the chart 
plt.scatter(y_axis,x_axis, c='red',label="Chicago")
plt.title("PyBar Fares per Month")
plt.legend()
plt.ylabel("Months")
plt.xlabel("Fares ($)")
plt.xlim(0,50) #set limit on x axis
plt.gca().invert_yaxis()
```

```python
#BUBBLE CHART useful for presenting financial, populate and weather data because you can add third and forth factor.  
# x_axis, y_axis, dot to bubble for size and color
```

```python
#change size of markers - use 's' s must equal feature of the data. s=y_axis
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
#Adjust the size of marker by multiplying the data in y_axis by any value. Iterate through each data point and multiply by 
# 3.
y_axis_larger = [] 
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
#List comprehension Python technique for efficiency when writing a 'for' loop. Optimized for Python to spot a 
# predictable pattern during looping
# for loop sets the larger size for the bubbles for readability
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Scatter Plot using the object-oriented interface
fig,ax= plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#object oriented interface bubble chart with 's' parameter for size
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Skill challenge 5.1.7
# s= list comprehension for loop to make bubble 5 times larger
# c = html hex color code (# must be added to code)
# alpha = transparency level, edgecolor= color ring around the bubble, linewidth = width of ring around bubble
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 7 for i in y_axis], c="#33C1FF", alpha=0.8, edgecolor = "black", linewidth=2,
          label= "Boston")
ax.set_title ("Pybar Fares by Month")
ax.set_xlabel ("Fares($)")
ax.set_xlim(0, 50)
ax.set_ylabel("Date")
ax.invert_yaxis()  #inverts y axis (jan to dec)
```
