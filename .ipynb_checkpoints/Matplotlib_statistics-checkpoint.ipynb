```python
#Import dependencies
#statistics module = add y-error bars and standard deviation
#Standard Deviation = measure of the amount of variation or spread of a set of values from the mean/average
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
#ADD DATASET
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#STANDARD DEVIATION
#Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
plt.errorbar(x_axis, y_axis, yerr=stdev,capsize=3)
```

```python
#STANDARD DEVIATION OBJECT ORIENTED METHOD
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#BAR CHARTS WITH ERROR BARS
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#BAR CHARTS WITH ERROR BARS OBJECT ORIENTED
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=5)
plt.show()
```

```python
#ADD MAJOR TICKS
#import numpy as np was added above with dependencies
#arrange = numpy arrange funtion, 0,51 = lower/upper limits, step = increments (5) to increase for each tick
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
#ADD MAJOR TICKS OBJECT ORIENTED 
#set_xticks() indicates x_axis, set_yticks() is y_axis
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#ADD MINOR TICKS 
#figsize = 8 X 8 in inches, barh = horizontal bar chart
#xticks = arrage low limit, high limit increment by 5
#multiplelocator = how often minor ticks. Minor ticks for every $1 on x_axis
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
